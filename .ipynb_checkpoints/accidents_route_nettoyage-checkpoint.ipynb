Certains horaires contiennent un point-virgule pour séparer les heures et minutes, d'autres pas. regardons d'abord les horaires avec deux-points.

```python
# Longueur des chaines de caractères correspondantes
data_caracteristiques.loc[(data_caracteristiques['hrmn'].str.contains(':')),'hrmn'].map(lambda x: len(x)).unique()
```

```python
# Longueur des chaines de caractères correspondantes
data_caracteristiques.loc[~(data_caracteristiques['hrmn'].str.contains(':')),'hrmn'].map(lambda x: len(x)).unique()
```

Les valeurs ont entre 1 et 4 caractères.

```python
# Valeurs à deux caractères
data_caracteristiques.loc[(~data_caracteristiques['hrmn'].str.contains(':')) & (data_caracteristiques['hrmn'].str.len()==2),'hrmn'].sort_values().unique()
```

```python
# Min et max des valeurs à trois caractères
print(data_caracteristiques.loc[(~data_caracteristiques['hrmn'].str.contains(':')) & (data_caracteristiques['hrmn'].str.len()==3),'hrmn'].min())
print(data_caracteristiques.loc[(~data_caracteristiques['hrmn'].str.contains(':')) & (data_caracteristiques['hrmn'].str.len()==3),'hrmn'].max())
```

```python
# Min et max des valeurs à quatre caractères
print(data_caracteristiques.loc[(~data_caracteristiques['hrmn'].str.contains(':')) & (data_caracteristiques['hrmn'].str.len()==4),'hrmn'].min())
print(data_caracteristiques.loc[(~data_caracteristiques['hrmn'].str.contains(':')) & (data_caracteristiques['hrmn'].str.len()==4),'hrmn'].max())
```

**Vérification des valeurs de la variable *dep*.**

```python
data_caracteristiques['dep'].sort_values().unique()
```

Là-aussi, il y a des problèmes à traiter. Commençons par les départements à un seul chiffre.

```python
data_caracteristiques.loc[data_caracteristiques['dep'].str.len()==1, 'dep'].unique()
```

```python
for i in range(1,10):
    start = data_caracteristiques.loc[data_caracteristiques['dep']==str(i), 'com'].map(lambda x : True if x.startswith(f"0{i}") else False)
    if start.all():
        print("dép numéro", i, ": communes commencent toutes par", f"0{i}")
    else:
        print("dép numéro", i, ": Problème -> communes ne commencent pas toutes par", f"0{i}")
```

Les départements 1, 2,...,9 peuvent être recodés en '01, '02',..., '09'.

```python
# Placer un 0 devant les numéros de dép à un seul chiffre
data_caracteristiques.loc[data_caracteristiques['dep'].str.len()==1, 'dep'] = (data_caracteristiques.loc[data_caracteristiques['dep'].str.len()==1, 'dep']
                                                                               .map(lambda x: '0' + x)
                                                                              )
```

Regardons maintenant les département à 3 chiffres et ne correspondent pas à l'Outre-Mer.

```python
#Numéros dép outre-mer
OM = data_caracteristiques['dep'].isin([str(i) for i in range(971, 990)])
#Numéros de départements de la table n'étant pas d'outre-mer mais ayant 3 chiffres
data_caracteristiques.loc[(~OM) & (data_caracteristiques['dep'].str.len()==3), 'dep'].sort_values().unique()
```

```python
data_caracteristiques.loc[data_caracteristiques['dep']=='100', 'com']
```

Les numéros de communes correspondant à ces départements sont aussi surprenants.

```python
data_caracteristiques.loc[(~OM) & (data_caracteristiques['dep'].str.len()==3), 'an'].sort_values().unique()
```

```python
data_caracteristiques.loc[(~OM) & ~(data_caracteristiques['dep'].str.len()==3), 'an'].sort_values().unique()
```

Les années pour lesquelles ces numéros de départements existent sont celles avant 2019. A partir de 2019, les seuls numéros de départements ayant 3 chiffres sont ceux de l'Outre-Mer.

En cas d'étude par département, il faudra faire attention à ces numéros problématiques.

**Vérification des valeurs de la variable *com*.**

Nous savons déjà que des valeurs posent problème. D'après la notice qui accompagne le jeu de données, les communes sont codées par 5 caractères dont les deux premiers sont ceux du département. Nous avons vu que ce n'est pas toujours le cas. Regardons s'il y a des années sans problèmes à ce niveau.

```python
data_caracteristiques.loc[data_caracteristiques['com'].str.len()==5, 'com']
```

Parmi les codes à 5 caractères, il y en a qui ont des caractères autre que chiffres ou lettres. Regardons uniquement ceux qui ont un code valide (5 caractères, uniquement des chiffres sauf le deuxième caractère qui peut-être un chiffre ou 'A' ou 'B')

```python
com_correcte = data_caracteristiques['com'].str.match(r'^\d[A|B|\d]\d{3}$').fillna(False)
data_caracteristiques.loc[~com_correcte, 'an'].sort_values().unique()
```

Là-encore, il n'y a a priori des problèmes que pour les années antérieures à 2019.

**Vérification des valeurs des variables *lat* et *long*.**

Le séparateur partie décimale/partie entière est une virgule. Pour passer les variables au format float, il faut la remplacer par un point.

```python
# Remplacement des virgules par des points dans lat et long pour transformation en type float
data_caracteristiques['lat']=data_caracteristiques['lat'].str.replace(',','.')
data_caracteristiques['long']=data_caracteristiques['long'].str.replace(',','.')
```

```python
# Retirer les espaces en début et fin de chaine
data_caracteristiques['lat']=data_caracteristiques['lat'].str.strip()
data_caracteristiques['long']=data_caracteristiques['long'].str.strip()
```

Les variables 'lat' et 'long' ne contiennent-elles que des valeurs numériques ?

```python
# sélection sans les valeurs manquantes
lat_non_manq = data_caracteristiques['lat'].dropna()
long_non_manq = data_caracteristiques['long'].dropna()
# vérification que les valeurs sont toutes de type numériques
print(lat_non_manq.str.match(r'^-?\d+(\.\d+)?$').all())
print(long_non_manq.str.match(r'^-?\d+(\.\d+)?$').all())
```

Les latitutdes sont toutes au format numérique.
Certaines longitudes n'ont pas un format numérique. Regardons ce qu'il en est.

```python
long_non_manq.loc[~long_non_manq.str.match(r'^-?\d+(\.\d+)?$')]
```

Ces valeurs de longitudes ne sont pas exploitables. A quelles années correspondent-elles ?

```python
(data_caracteristiques.dropna(subset='long')
                      .loc[~long_non_manq.str.match(r'^-?\d+(\.\d+)?$'), 'an']
                      .unique())
```

Ces longitudes non exploitables correspondent uniquement à des données de 2009. Il faudra en tenir compte en cas d'utilisation de la variable.

### Étude des valeurs manquantes

```python
import missingno as msno
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches  #module qui permet de construire une légende pour le graph

# Utilisation de la police de base de Latex pour harmonisation des polices dans le rapport
plt.rc('font', family = 'serif', serif = 'cmr10')
plt.rcParams.update({"text.usetex": True, "axes.formatter.use_mathtext" : True})

# Graphe
msno.matrix(data_caracteristiques.sort_values('an'))
# Création de la légende
gray_patch = mpatches.Patch(color='dimgray', label='Données présentes')
white_patch = mpatches.Patch(facecolor='white', edgecolor='dimgray', label='Données absentes')
plt.legend(handles=[gray_patch, white_patch], bbox_to_anchor=(0,0.5), fontsize=14)

#Titre et affichage
plt.title("Données manquantes pour chaque variable du jeu de données", fontsize=18)
plt.show()
```

- La présence des valeurs de *lat* semble parfaitement corrélée à celle des valeurs de *long*.
- La présence de l'adresse semble invrsement corrélée avec celle de *lat* et *long*. vérifions par une table de corrélation.

```python
msno.heatmap(data_caracteristiques)
plt.show()
```

- les valeurs manquantes de *lat* et *long* sont parfaitement corrélées.
- En revanche, *adr* possède une corrélation faible avec *lat* et *long*.

Regardons les valeurs manquantes uniquement à partir de 2019 (nous avons vu plus haut que les données sont souvent de meilleure qualité à partir de 2019).

```python
# Graphe
msno.matrix(data_caracteristiques.loc[data_caracteristiques['an']>=2019].sort_values('an'))
# Création de la légende
gray_patch = mpatches.Patch(color='dimgray', label='Données présentes')
white_patch = mpatches.Patch(facecolor='white', edgecolor='dimgray', label='Données absentes')
plt.legend(handles=[gray_patch, white_patch], bbox_to_anchor=(0,0.5), fontsize=14)

#Titre et affichage
plt.title("Données manquantes pour chaque variable du jeu de données", fontsize=18)
plt.show()
```

```python
data_caracteristiques.loc[data_caracteristiques['an']>=2019].isna().sum()
```

Très peu de valeurs manquantes à partir de 2019.

**Vérification des valeurs de la variable *secu***

```python
data_usagers['secu'].sort_values().unique()
```

La variable secu des fichiers d'avant 2019 correspond à la variable secu1 post-2019. Le premier chiffre correspond au même codage de l'équipement de sécu que secu1. Le deuxième chiffre indique si l'équipement était utilisé. Dans secu1, l'équipement est indiqué uniquement s'il est utilisé. Sinon, on indique le code 0. Pour toutes les lignes pré-2019, codons dans secu1 la valeur correspondante de secu.

```python
conditions = [(data_usagers['secu']//10!=0) & (data_usagers['secu']%10==1),
              (data_usagers['secu']//10!=0) & (data_usagers['secu']%10!=1) & (~data_usagers['secu'].isna())
             ]

resultats = [data_usagers['secu']//10,
             0
            ]

data_usagers['secu1'] = np.select(conditions, resultats, data_usagers['secu1'])
data_usagers.loc[:,['secu','secu1']]
```

La variable *secu* ne servira plus. On peut la supprimer.

```python
data_usagers.drop(columns=['secu'], inplace=True)
```

## Étude des valeurs manquantes

```python
msno.matrix(data_usagers)
plt.show()
```

- Beaucoup de valeurs manquantes pour *secu2* et *secu3* mais cela est normal compte-tenu qu'il s'agit de variables identifiants un deuxième ou un troisième équipement de sécurité utilisé sur un véhicule. Il y a moins de valeurs manquantes pour *secu1*.
- Il y a une certaine corrélation entre valeurs manquantes de locp et actp. C'est normal aussi car ces variables concernent les piétons toutes les deux. Il est aussi normal d'avoir beaucoup de valeurs manquantes pour etatp car cette variable ne contient pas de valeur 0 s'il n'y a pas de piéton impliqué dans l'accident, contrairement à *locp* et *actp*.
- Là-encore, beaucoup de valeurs manquantes pour l'identifiant du véhicule.

Regardons ce que cela donne pour les accidents à partir de 2019.

```python
us_post_2019 = (data_caracteristiques
                .loc[data_caracteristiques['an']>=2019, ['Num_Acc']]
                .merge(data_usagers, how='inner', left_on='Num_Acc', right_on='Num_Acc')
               )

msno.matrix(us_post_2019)
plt.show()
```

- Il n'y a presque plus d'identifiant de véhicule manquant.
- A part les variables qui ont naturellement des valeurs manquantes, les autres ont très peu de valeurs manquantes.
