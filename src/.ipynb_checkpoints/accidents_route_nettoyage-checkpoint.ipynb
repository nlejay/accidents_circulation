# Nettoyage des données des fichiers sur les accidents de la route entre 2005 et 2021

***
## Fichier caracteristiques_2005_2021.csv



```python
import pandas as pd

data_caracteristiques = pd.read_csv('caracteristiques_2005_2021.csv', sep=';', encoding='utf-8',
                                   dtype={'hrmn': 'str', 'com' : 'str', 'lat':'str', 'long':'str', 'dep':'str'})
```

```python
data_caracteristiques.head()
```

```python
data_caracteristiques.info()
```

### Recherche de doublons

```python
data_caracteristiques.duplicated(subset=['Num_Acc']).sum()
```

### Changement du type des variables et correction de certaines valeurs



```python
# 'Num_Acc' : passage en type str
data_caracteristiques['Num_Acc'] = data_caracteristiques['Num_Acc'].astype('str')
```

```python
# atm est de type float et devrait être de type int
# Recherche des valeurs pour vérification avant passage en type int
data_caracteristiques['atm'].unique()
```

```python
data_caracteristiques['atm'] = data_caracteristiques['atm'].astype('Int64')
```

```python
# Même chose avec col
data_caracteristiques['col'].unique()
```

```python
data_caracteristiques['col'] = data_caracteristiques['col'].astype('Int64')
```

```python
# Remplacement des virgules par des points dans lat et long pour transformation en type float
data_caracteristiques['lat']=data_caracteristiques['lat'].str.replace(',','.')
data_caracteristiques['long']=data_caracteristiques['long'].str.replace(',','.')
```

```python
# Certaines longitudes doivent être négatives
data_caracteristiques.loc[(~data_caracteristiques['long'].isna())&(data_caracteristiques['long'].str.contains("-"))]
```

```python
# Certaines longitudes ne contiennent que le symbole -
data_caracteristiques.loc[data_caracteristiques['long']=="-",'long']
```

```python
# Remplacement de - par NaN
import numpy as np
data_caracteristiques.loc[data_caracteristiques['long']=="-",'long']=np.NaN
```

```python
# Passage de lat et long en type float
data_caracteristiques['lat']=data_caracteristiques['lat'].astype('Float64')
data_caracteristiques['long']=data_caracteristiques['long'].astype('Float64')
```

```python
data_caracteristiques['dep']=data_caracteristiques['dep'].astype('str')
```

```python
data_caracteristiques.info()
```

Toutes les variables ont le bon format.

### Etude des valeurs manquantes

```python
data_caracteristiques.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 75
import matplotlib.patches as mpatches  #module qui permet de construire une légende pour le graphe
import missingno as msno

# Représentation de la matrice des valeurs manquantes
msno.matrix(data_caracteristiques.sort_values('lat'))

# Création de la légende
gray_patch = mpatches.Patch(color='dimgray', label='Données présentes')
white_patch = mpatches.Patch(facecolor='white', edgecolor='dimgray', label='Données absentes')
plt.legend(handles=[gray_patch, white_patch], bbox_to_anchor=(0,0.5), fontsize=14)
plt.title("Données manquantes pour chaque variable du jeu de données", fontsize=18)

# Affichage
plt.show()
```

```python
data_caracteristiques.loc[(data_caracteristiques['lat'].astype('str').str.len()==3) & (data_caracteristiques['long'].astype('str').str.len()==3), ['lat','long']] = data_caracteristiques.loc[(data_caracteristiques['lat'].astype('str').str.len()==3) & (data_caracteristiques['long'].astype('str').str.len()==3), ['lat','long']]= np.NaN
```

Toutes les heures non saisies avec le symbole ":" et qui commencent par des 0 ont perdu ces 0.

```python
# Pour chaque valeur de longueur inférieure à 5 dans la colonne hrmn, 
# on ajoute autant de 0 que nécessaire au début de la chaine puis le reste de la chaine avec ":" avant les deux derniers chiffres
for i in range(1,5):
    data_caracteristiques.loc[data_caracteristiques['hrmn'].str.len()==i,'hrmn']=("0"*(4-i)+
                                                                                  data_caracteristiques.loc[data_caracteristiques['hrmn'].str.len()==i,'hrmn'].str[:-2]+
                                                                                  ":"+
                                                                                  data_caracteristiques.loc[data_caracteristiques['hrmn'].str.len()==i,'hrmn'].str[-2:])
```
